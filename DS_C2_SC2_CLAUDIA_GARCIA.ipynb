DATAFRAME NUEVO LLAMADO "CON DIFERENCIAS"

```python
import pandas as pd
vacunas = pd.read_csv('/content/drive/My Drive/Colab Notebooks/country_vaccinations.csv')
vacunas['diferencias'] = vacunas['total_vaccinations'] - vacunas['people_vaccinated']

con_diferencias = vacunas.groupby('country')['diferencias'].sum().reset_index()
display(con_diferencias)
```

OBTENER LA DIFERENCIA DE TIEMPO

```python
import pandas as pd
vacunas = pd.read_csv('/content/drive/My Drive/Colab Notebooks/country_vaccinations.csv')


vacunas['date'] = pd.to_datetime(vacunas['date'], dayfirst=True)

fechas_por_pais = vacunas.groupby('country')['date'].agg(['min', 'max'])

fechas_por_pais['diferencia_dias'] = fechas_por_pais['max'] - fechas_por_pais['min']

display(fechas_por_pais)
```
